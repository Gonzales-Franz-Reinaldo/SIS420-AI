```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generar datos sintéticos con 3 clústeres
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', edgecolor='black', s=50)
plt.title("Datos sin etiquetas")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Inicialización: Seleccionar aleatoriamente 3 centroides iniciales
k = 3
kmeans = KMeans(n_clusters=k, init='random', random_state=42)

# Asignación de clústeres y actualización iterativa
kmeans.fit(X)

# Obtener las etiquetas de clúster asignadas y los centroides finales
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(8, 6))

# Visualizar los puntos de datos agrupados por clúster
for cluster in range(k):
    plt.scatter(X[cluster_labels == cluster, 0], X[cluster_labels == cluster, 1], 
                label=f'Cluster {cluster}', s=50)

# Visualizar los centroides
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red', s=200, label='Centroides')

plt.title("Clustering con K-means")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
