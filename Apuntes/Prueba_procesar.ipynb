```python
# Hacer una copia del DataFrame
data = df.copy()

# Reemplazar 'cancelado' por 1 y 'no_cancelado' por 0 en la columna 'loan_status'
data['loan_status'] = data['loan_status'].replace({'cancelado': 1, 'no_cancelado': 0})

# Mostrar los datos
data
```

```python
# mostramos cuantas clases tiene la columna de loan_status
data['loan_status'].value_counts()
```

```python
#variables categoricas, donde convertimos a int64
columnas_categoricas = data.select_dtypes(include=['object']).columns

#Procesamiento de frases o palabras a valores numericos para su procesamiento
for columna in columnas_categoricas:
  le = LabelEncoder()
  data[columna] = le.fit_transform(data[columna])
  
# Mostramos los datos
data
```
