## Para eliminar Columnas 

Eliminamos las columnas que no aportan nada al proceso de entrenamiento

```python
#Descartamos las columnas que tienen muchos datos perdidos o que no son muy necesarios para el analisis
data.drop(['Type'], axis=1, inplace=True)
data.drop(['Current_Ver'], axis=1, inplace=True)
data.drop(['Android_Ver'], axis=1, inplace=True)
```

## Completar los campos vacios de las columnas

Como vemos que las columnas Rating, Current_Ver y Android_Ver tienen datos faltantes, procedemos a llenarlos con la media de los datos de la columna

```python
#Sacamos los valores minimos y maximos de cada columna que necesite rellenar los datos
# Ojo que tiene que ser columnas de puro numeros

# OTRO METODO
print("Rating")
print("Valor minimo: ", data['Rating'].min())
print("Valor máximo: ", data['Rating'].max())

print("Current_Ver")
print(data['Current_Ver'].min())
print(data['Current_Ver'].max())
```

```python
#Reemplazamos todos los valores NaN de las columnas que tenian campos faltantes

# Para numpy
for i in range(len(data)):
    if pd.isna(data.at[i, 'Rating']):
        data.at[i, 'Rating'] = np.random.randint(1, 5)

#? ==================================================================
# Con pandas
# Sacamos los valores mínimos y máximos de la columna 'Rating'
rango_min = data['Rating'].min()
rango_max = data['Rating'].max()

# Reemplazamos los valores NaN en la columna 'Rating' con los valores mínimo y máximo
data['Rating'].fillna(np.random.randint(rango_min, rango_max + 1), inplace=True) # puede ser esto 1

data['Rating'] = data['Rating'].fillna(np.random.randint(rango_min, rango_max + 1))  # puede ser esto 2

data.loc[:, 'Rating'] = data['Rating'].fillna(np.random.randint(rango_min, rango_max + 1)) # puede ser esto 3


# Verificamos si hay valores NaN en la columna 'Rating' después del reemplazo
print("Número de valores NaN en la columna 'Rating' después del reemplazo:", data['Rating'].isna().sum())
```

```python
# Verificamos de nuevo la información del dataset para ver si hay columnas con campos vacios
print('\nDATOS VACIOS')
print(pd.isnull(data).sum())
```

## Convertir todas las columnas string a números 

```python
#labelEncoder sirve para convertir datos categoricos a numericos para poder hacer mejores calculos
from sklearn.preprocessing import LabelEncoder

#variables categoricas, donde convertimos a int64
columnas_categoricas = data.select_dtypes(include=['object']).columns

#Procesamiento de frases o palabras a valores numericos para su procesamiento
for columna in columnas_categoricas:
  le = LabelEncoder()
  data[columna] = le.fit_transform(data[columna])
  

# Verificamos el dataset
data.head(5)
```

## Para mover columnas al principio o al final

```python
#Movimiento de columna al final del dataset y verificacion de datos
# En este caso movemos nuestra Y predicha al final de las columnas que en este caso es "Install"
indice_columna = data.columns.get_loc('Installs')
columnas = list(data.columns)
columnas.pop(indice_columna)
columnas.append('Installs')
data = data[columnas]

# Verificamos el dataset
data.head(5)
```
